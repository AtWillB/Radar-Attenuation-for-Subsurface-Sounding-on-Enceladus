```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.transforms as mtransforms


import radar_attenuation.depth_map_util as dmu

cmap = sns.color_palette("rocket_r")
```

```python
hemingway_d_depth_map = np.loadtxt("../../figures/figure3/data/ice_shell_depth_maps/Hemingway_depth_map_b.csv", delimiter= ',')
regolith_depth_dict = dmu.create_regolith_maps()
```

```python
filepath = "../../figures/figure3/data/attenuation_calculations_maps/low_loss/hemingway_map_B_bc_relative_low_loss.csv"
hemingway_low_d_bc_reg_depthpen =  np.loadtxt(filepath, delimiter=',')

filepath = "../../figures/figure3/data/attenuation_calculations_maps/low_loss/hemingway_map_B_exp_relative_low_loss.csv"
hemingway_low_d_exp_reg_depthpen = np.loadtxt(filepath, delimiter=',')

filepath = "../../figures/figure3/data/attenuation_calculations_maps/low_loss/hemingway_map_B_no_relative_low_loss.csv"
hemingway_low_d_no_reg_depthpen =  np.loadtxt(filepath, delimiter=',')
```

```python
filepath = "../../figures/figure3/data/attenuation_calculations_maps/high_loss/hemingway_map_B_bc_relative_high_loss.csv"
hemingway_high_d_bc_reg_depthpen = np.loadtxt(filepath, delimiter=',')

filepath = "../../figures/figure3/data/attenuation_calculations_maps/high_loss/hemingway_map_B_exp_relative_high_loss.csv"
hemingway_high_d_exp_reg_depthpen =np.loadtxt(filepath, delimiter=',')

filepath = "../../figures/figure3/data/attenuation_calculations_maps/high_loss/hemingway_map_B_no_relative_high_loss.csv"
hemingway_high_d_no_reg_depthpen = np.loadtxt(filepath, delimiter=',')
```

```python
x, y = dmu.create_coords(181, 361)
fig, ax = plt.subplots(3, 3, figsize = (39, 15), dpi = 500)
label_fontsize = 18
tick_fontsize = label_fontsize - 3
x_label = 'Longitude [deg]'
y_label = 'Latitude [deg]'
limit = 100
calc_layer_num = 100
rocket_colors = sns.color_palette('rocket', 99) + [(0.9955451416666666, 0.97512059, 0.960499675)]
rocket_cmap = ListedColormap(sns.color_palette(rocket_colors, as_cmap=True))


dmu.mk_depth_map(hemingway_d_depth_map/1000, x, y, "cividis", 'Ice Shell Depth [km]', 7, fig, tick_fontsize, ax = ax[0, 0], depth_map_limit = np.array(hemingway_d_depth_map/1000).max())
dmu.depth_map_formatting(ax[0, 0], y_ticks=True, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_low_d_no_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize, ax = ax[0, 1], atten_limit = limit)
dmu.depth_map_formatting(ax[0, 1], cbar, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_high_d_no_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize, ax = ax[0, 2], atten_limit = limit)
dmu.depth_map_formatting(ax[0, 2], cbar, label_fontsize = 18)
ax[0, 1].text(-40, 95, s = 'Low Loss', fontweight = 'bold', fontsize = label_fontsize+10)
ax[0, 2].text(-40, 95, s = 'High Loss', fontweight = 'bold', fontsize = label_fontsize+10)



cbar = dmu.mk_depth_map(regolith_depth_dict['exp']/1000, x, y, "cividis", 'Regolith Depth [km]', 20, fig, tick_fontsize, ax = ax[1, 0], depth_map_limit =  np.array(regolith_depth_dict['exp']/1000).max())
dmu.depth_map_formatting(ax[1, 0], cbar, y_ticks=True, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_low_d_exp_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize, ax = ax[1, 1], atten_limit = limit)
dmu.depth_map_formatting(ax[1, 1], cbar, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_high_d_exp_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize, ax = ax[1, 2], atten_limit = limit)
dmu.depth_map_formatting(ax[1, 2], cbar, label_fontsize = 18)


cbar = dmu.mk_depth_map(regolith_depth_dict['bc']/1000, x, y, "cividis", 'Regolith Depth [km]', 20, fig, tick_fontsize, ax = ax[2, 0], depth_map_limit = 0.5)
dmu.depth_map_formatting(ax[2, 0], cbar, y_ticks = True, x_ticks= True, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_low_d_bc_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize,ax = ax[2, 1], atten_limit = limit)
dmu.depth_map_formatting(ax[2, 1], cbar, x_ticks= True, label_fontsize = 18)
cbar = dmu.mk_depth_map(hemingway_high_d_bc_reg_depthpen, x, y, rocket_cmap, '2-Way Attenuation [dB]', calc_layer_num, fig, tick_fontsize,ax = ax[2, 2], atten_limit = limit)
dmu.depth_map_formatting(ax[2, 2], cbar, x_ticks= True, label_fontsize = 18)

# Colorbar for both exponential and best case regolith distributions
norm = plt.Normalize(0, 700)
cmap = plt.get_cmap('cividis', 20)
sm = plt.cm.ScalarMappable(cmap= cmap, norm=norm)
sm.set_array([])
cbar_ax = [0.326, 0.1, .0055, .525]
cbar_ax = fig.add_axes(cbar_ax)
cb = fig.colorbar(sm, orientation="vertical", pad=0.02, shrink=0.8, cax = cbar_ax)
cb.set_label("Porous Layer Thickness [m]", fontsize = tick_fontsize)
cb.set_ticklabels([0, 100, 200, 300, 400, 500, 600, 700], fontsize = tick_fontsize-2)


cbar_ax = [0.595, 0.1, .006, .8]
dmu.mk_custom_cbar(cbar_ax, fig, label_fontsize=18)
cbar_ax = [0.86, 0.1, .006, .8]
dmu.mk_custom_cbar(cbar_ax, fig, label_fontsize = 18)


# plt.tight_layout()
plt.subplots_adjust(left= 0.1,
                    bottom = 0.1, 
                    right = 0.9, 
                    top = 0.9, 
                    wspace = 0, 
                    hspace=0.1)

label_list  = ["a)", "b)", "c)", "d)", "e)", "f)", "g)", "h)", "i)"]
for label, ax in zip(label_list, ax.flatten()):
    if label in ("d)", "g)"):
        color = "white"
    else:
        color = "black"
    trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
    ax.text(0.02, 0.98, label, transform=ax.transAxes + trans,
            fontsize= label_fontsize+6, verticalalignment='top', fontfamily='serif',color = color,
            bbox=dict(facecolor='none', edgecolor = "none", pad=3.0))
plt.savefig("../../figures/figure3/image(s)/fig3b.png", bbox_inches='tight')
```
