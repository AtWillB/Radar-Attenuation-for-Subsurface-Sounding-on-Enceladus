```python
#load up the appropriate ice shell depth map and depth_to_100_dB attenuation map, and plug them into create_depth_pen
path_to_low_a_exp_attenuation_map = "../../figures/figure3/data/attenuation_calculations_maps/low_loss/hemingway_map_A_exp_relative_low_loss.csv"
hemingway_low_a_exp_reg_depthpen = np.loadtxt(path_to_low_a_exp_attenuation_map, delimiter= ',')



path_to_high_b_bc_attenuation_map = "../../figures/figure3/data/attenuation_calculations_maps/high_loss/hemingway_map_B_exp_relative_high_loss.csv"
hemingway_high_b_bc_reg_depthpen = np.loadtxt(path_to_high_b_bc_attenuation_map, delimiter= ',')
```
